## Ekstra - Udvid Matplotlib med Seaborn

Seaborn er et ekstra bibliotek til Matplotlib, der giver enklere måder at lave visuelle plots. De har en masse forskellige plots, som vi kan bruge til at visualisere data. Lad os tage et lille eksempel.

### Heatmap
Heatmap er en grafisk repræsentation af data, hvor værdierne er repræsenteret ved farve. Det er en nyttig måde at visualisere data på, da det giver en hurtig oversigt over dataene.


#### Heatmap over gennemsnitlige kvadratmeter pris for hver by

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = './Data/MatPlotLib-Data/DKHousingPricesSample100k.csv'
housing_data = pd.read_csv(file_path)

# Prepare data for the heatmap
# Group by city and calculate the average sqm price
heatmap_data = housing_data.groupby('city')['sqm_price'].mean().sort_values(ascending=False)

# Select the top 50 cities with the highest average sqm price
heatmap_data = heatmap_data.head(50).reset_index()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    heatmap_data.set_index('city'),  # Transpose to use cities as columns for better visualization
    annot=True, 
    fmt='.0f', 
    cmap='coolwarm', 
    cbar_kws={'label': 'Average Sqm Price (DKK)'}
)
plt.title('Top 20 Cities by Average Sqm Price', fontsize=16)
plt.ylabel('City', fontsize=12)
plt.xlabel('Average Sqm Price', fontsize=12)
plt.tight_layout()
plt.show()
````

#### Heatmap over korrelationen mellem forskellige karakteristika i brugt bil data

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the uploaded file to examine its structure
file_path = './Data/MatPlotLib-Data/car_prediction_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows and general info to understand the data structure
data.head(), data.info()

# Select only numeric columns for correlation
numeric_columns = ['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven', 'Owner']
correlation_matrix = data[numeric_columns].corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
````

#### Heatmap Analyse - Brugtbilsdata

##### Negative Korrelationer
- Der er en moderat negativ korrelation (-0.52) mellem `Year` og `Kms_Driven`, hvilket giver mening da ældre biler typisk har kørt flere kilometer.
- `Year` har også en svag negativ korrelation (-0.18) med `Owner`, hvilket antyder at ældre biler oftere har haft flere ejere.

##### Svage Korrelationer
- `Owner` har generelt svage korrelationer med de andre variable (værdier tæt på 0), hvilket indikerer at antallet af tidligere ejere ikke har stor indflydelse på de andre faktorer.
- `Kms_Driven` har en meget svag korrelation med `Selling_Price` (0.03) og `Present_Price` (0.20), hvilket er interessant da man kunne forvente en stærkere negativ korrelation.

##### Praktiske Implikationer
- Ved prissætning af brugte biler bør man primært fokusere på bilens oprindelige pris (`Present_Price`).
- Bilens alder (`Year`) er også en vigtig faktor at tage i betragtning.
- Antallet af tidligere ejere ser ud til at have mindre betydning for prissætningen.

Denne information kan være nyttig for både købere og sælgere på brugtbilsmarkedet, samt for udvikling af prisestimerings-modeller.

### Boxplot
Boxplot er en grafisk repræsentation af data, der viser fordelingen af dataene. Det er en nyttig måde at visualisere data på, da det giver en hurtig oversigt over dataene.

##### Boxplot består af 5 elementer:
 1. Minimum: Den laveste værdi i datasættet (nederste whisker)
 2. Først kvartil (Q1): 25% af dataene ligger under denne værdi (bunden af boksen)
 3. Median (2. kvartil): Den midterste værdi, hvor 50% ligger over og under (linjen i boksen)
 4. Tredje kvartil (Q3): 75% af dataene ligger under denne værdi (toppen af boksen)
 5. Maximum: Den højeste værdi i datasættet (øverste whisker)


````python
import pandas as pd

car_prediction_data = pd.read_csv('./Data/MatPlotLib-Data/car_prediction_data.csv')
# Filter out rows where Fuel_Type is 'CNG'
filtered_car_data = car_prediction_data[car_prediction_data['Fuel_Type'] != 'CNG']

# Create a new boxplot for selling prices of cars, grouped by fuel type (CNG removed)
filtered_car_data.boxplot(column='Selling_Price', by='Fuel_Type', grid=False, showfliers=False)

# Set the title and labels for better visualization
plt.title('Selling Price Distribution by Fuel Type (CNG Removed, Outliers Excluded)')
plt.suptitle('')  # Remove the automatic subtitle from pandas
plt.xlabel('Fuel Type')
plt.ylabel('Selling Price (Lakhs)')

# Display the boxplot
plt.show()
````

##### Boxplot Analyse - Brændstoftyper og Salgspriser

Prisniveau
- Dieselbiler har generelt en højere salgspris end benzinbiler
- Medianprisen (den midterste streg i boksen) for dieselbiler ligger omkring 8-9 lakhs
- Medianprisen for benzinbiler ligger omkring 3-4 lakhs

Prisspredning
- Dieselbiler har en større prisspredning (større boks og længere whiskers)
- Prisintervallet for dieselbiler strækker sig fra ca. 3 til 23 lakhs
- Benzinbiler har en mere kompakt prisfordeling, primært mellem 0.5 og 11 lakhs

Praktiske Implikationer
- Hvis man leder efter en bil i det lavere prissegment, er benzinbiler typisk det bedste valg
- Dieselbiler tilbyder flere valgmuligheder i det højere prissegment
- Der er større prisvariation blandt dieselbiler, hvilket kan skyldes flere forskellige modeltyper eller årgange
